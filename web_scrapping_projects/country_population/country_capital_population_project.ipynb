```python
import requests, pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.worlddata.info/capital-cities.php"
page = requests.get(url)
page.status_code
```

```python
soup = BeautifulSoup(page.text, "html.parser")
```

```python
# table header
table = soup.find_all('table')[0]
header = [title.text for title in table.find_all('th')]
df = pd.DataFrame(columns=header)
```

```python
# table rows
for table in soup.find_all('table'):
    for rows in table.find_all('tr')[1:]:
        ind_row = [row_data.text for row_data in rows]
        
        length = len(df)
        df.loc[length] = ind_row
    
    
```

```python
df.to_csv("country_capital_population.csv", index=False)
```

```python
def get_column(df):
    if df['Population'].find(','):
        return df['Population'].replace(',', '')
    else:
        return df['Population']
```

```python
df['Population'] = df.apply(get_column, axis=1).astype('int64')
```
