# Pivot

### To show the evalution of gdp per country over the years

```python
# add "encoding='ISO-8859-1" to the read argument if unicode error occurs
df_gdp = pd.read_csv('data/gdp.csv', encoding='ISO-8859-1') 
```

```python
# reshpae the dataframe with .pivot()
df_gdp.pivot(index="year", columns="country", values="gdppc")
```

## 2 pivot_table()

```python
# read and show data set
df_sales = pd.read_excel('data/supermarket_sales.xlsx')
df_sales.head()
```

```python
# make a pivot table and add aggregate function
df_sales.pivot_table(index="Gender", values=['Quantity', 'Tax 5%', 'Total', 'Unit price', 'cogs',  'gross income', 'gross margin percentage'], aggfunc="sum")
#df_sales.info()
```

```python
# make a pivot table and add aggregate function and select some columns
df_sales.pivot_table(index="Gender", values=['Quantity', 'Total'], aggfunc="sum")
```

```python
# a pivot table that shows how much male and female spend in each product line
df_sales.pivot_table(index="Gender", columns="Product line", values='Total', aggfunc="sum")
```
