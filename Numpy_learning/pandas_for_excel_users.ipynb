# 2 Creating a DataFrame from a Dictionary

## 1 Value Counts

```python
# counting gender element

# len function
len(data['gender'])
# .count method
data['gender'].count()
```

```python
# counting gender elements by category
data['gender'].value_counts()
```

```python
# return the relative frequency (divide all values by the sum of values)
data['gender'].value_counts(normalize=True)
```

```python
# filtering data base on a column values
data[data['gender'] == 'female']
```

```python
# count "parental level of education" element by category
data['parental level of education'].value_counts()
```

```python
# return the relative frequency and round up to 2 decimal places
data['parental level of education'].value_counts(normalize=True).round(2)
```

# Sort a Dataframe
