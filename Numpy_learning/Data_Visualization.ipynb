```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_population = pd.read_csv('data/population_total.csv')
```

# Making a Pivot Table

### Evolution of countries population over the years

```python
# drop all null values from the dataframe
df_population.dropna(inplace=True)
```

```python
# making a pivot table
df_pivot = df_population.pivot(index='year', columns='country', values='population')
```

```python
# selecting some columns
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
# showing the pivot table
df_pivot.head()
```

# 2 LinePlot

```python
df_pivot.plot(kind='line', xlabel='Year', ylabel='Population', title='Population (1955 - 2020)', figsize=(8,4))
plt.savefig('population_lineplot.png')
plt.show()
```

```python
# export the pivot table to excel
df_pivot.to_excel('data/pivot_table.xlsx')
```

# 3 Barplot

```python
# selecting only one year 2020
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_2020
```

```python
# transpose dataframe from rows to column and viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# making the barplot
df_pivot_2020.plot(kind='bar', color='green', xlabel='Year', ylabel='Population', title='Population 2020')
plt.savefig('population_barchart.png')
plt.show()
```

# 3.2 Barplot grouped by "n" variables

```python
# select some years 
df_pivot_sample = df_pivot[df_pivot.index.isin([1980,1990,2000,2010,2020])]
```

```python
# making the grouped barplot
df_pivot_sample.plot(kind='bar', xlabel='Year', ylabel='Population')
```
