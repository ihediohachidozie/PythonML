# Read a .csv from a URL with Pandas
Target website: https://www.football-data.co.uk/data.php

```python
# reading 1 csv file from the website
df_premier24 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2324/E0.csv')
```

```python
# showing dataframe
df_premier24.head()
```

```python
# rename columns
df_premier24 = df_premier24.rename(columns={'Date':'date',
                                            'HomeTeam':'home_team',
                                            'AwayTeam':'away_team',
                                            'FTHG':'home_goals',
                                            'FTAG':'away_goals'})
```

```python
# show dataframe
df_premier24
```
