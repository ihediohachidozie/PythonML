```python
import numpy
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# mean value of the speed arry
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
x = numpy.mean(speed)
x
```

```python
# mediam value of the array
# 77, 78, 85, 86, 86, 87, 87, 88, 94, 99, 111, 103
y = numpy.median(speed)
y
```

```python
numbers = speed.copy()
x = 0

while(x < len(numbers)):
  y = x + 1
  while(y < len(numbers)):
    if numbers[x] < numbers[y]: # here determines the order (greater then ascending order & less than descending order)
      numbers[x], numbers[y] = numbers[y], numbers[x]
    y += 1
  x += 1
numbers
```
